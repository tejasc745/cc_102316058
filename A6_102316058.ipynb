Q1. Ask the user to enter a value (e.g., M) for a mathematical function.
Generate x values from -10 to 10 using np.linspace().
Compute y values for:
- y= M⋅x^2
- y=M⋅sin(x)

Plot both functions in a single figure:
* Use different colors and line styles.
* Add legend, grid, and title.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

M = float(input("Enter a value for M : "))


x = np.linspace(-10, 10, 100)


y1 = M * x**2
y2 = M * np.sin(x)


plt.figure(figsize=(10, 5))
plt.plot(x, y1, 'b--', label='y = M * x^2')
plt.plot(x, y2, 'r-', label='y = M * sin(x)')
plt.legend()
plt.xlabel('x values')
plt.ylabel('y values')
plt.title('Mathematical Functions Plot')
plt.grid(True)
plt.show()

```

Q.2 Create a dataset of five subjects and scores. Convert it into a Pandas DataFrame.
Plot the scores using a Seaborn bar plot with :
- Different colors for each bar.
- Annotations on top of each bar.
- Title, axis labels, and grid.


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

subjects = ['Math', 'Physics', 'Chemistry', 'Biology', 'English']
scores = [85, 78, 92, 88, 75]


df = pd.DataFrame({'Subjects': subjects, 'Scores': scores})

# Plot using Seaborn
plt.figure(figsize=(8, 5))
sns.barplot(x='Subjects', y='Scores', data=df, palette='viridis')

# Add annotations
for index, value in enumerate(scores):
    plt.text(index, value + 1, str(value), ha='center', fontsize=12)

# Labels and Title
plt.xlabel("Subjects")
plt.ylabel("Scores")
plt.title("Scores in Different Subjects")
plt.grid(axis='y')

plt.show()
```

Q3. Write a Python script to select your roll number as seed for NumPy and then generate a dataset of 50 values using np.random.randn().
Create a 2x2 subplot layout displaying:
- Line plot showing cumulative sum.
- Scatter plot with random noise.
Customize with titles, axis labels, and grids.

```python
import numpy as np
roll_number = 102316037
np.random.seed(roll_number)


data = np.random.randn(50)


fig, ax = plt.subplots(2, 2, figsize=(10, 8))


ax[0, 0].plot(np.cumsum(data), 'g-', label="Cumulative Sum")
ax[0, 0].set_title("Cumulative Sum Plot")
ax[0, 0].legend()
ax[0, 0].grid(True)


ax[0, 1].scatter(range(50), data, color='red', alpha=0.6)
ax[0, 1].set_title("Random Noise Scatter Plot")
ax[0, 1].grid(True)


plt.tight_layout()
plt.show()
```

Q.4 Download Data-set from the below link
https://github.com/AnjulaMehto/MCA/blob/main/company_sales_data.csv
Apply ‘seaborn’ library to do the following :
1. Read Total profit of all months and show it using a line plot.
2. Read all product sales data and show it using a multiline plot.
3. Plot bar chart for all the features/attributes.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

url = 'https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv'
df = pd.read_csv(url)


plt.figure(figsize=(8, 5))
plt.plot(df['month_number'], df['total_profit'], marker='o', linestyle='-', color='b', label='Total Profit')
plt.title("Total Profit Over Months")
plt.xlabel("Month Number")
plt.ylabel("Total Profit")
plt.legend()
plt.grid(True)
plt.show()


plt.figure(figsize=(8, 5))
for column in df.columns[1:-2]:
    plt.plot(df['month_number'], df[column], marker='o', linestyle='--', label=column)
plt.title("Product Sales Over Months")
plt.xlabel("Month Number")
plt.ylabel("Sales")
plt.legend()
plt.grid(True)
plt.show()

df.iloc[:, 1:].plot(kind='bar', figsize=(12, 6), width=0.8)
plt.title("All attributes")
plt.xlabel("Month Number")
plt.ylabel("Values")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

df.drop(columns=['month_number','total_units', 'total_profit']).plot(kind='bar', figsize=(12, 6), width=0.8)
plt.title("Company Sales Data Bar Chart")
plt.xlabel("Month Number")
plt.ylabel("Values")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
